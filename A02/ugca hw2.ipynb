```python
import nltk
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import wordnet
from nltk.corpus import stopwords
from nltk import FreqDist
```

```python
import re
```

```python
len(data)
```

```python
text=data["text"].values.tolist()
```

```python
text=" ".join(text)
```

```python
stops=stopwords.words('english')
```

```python
def clean_data(text):
    word=[]
    #delete punctuations stopwords
    words = re.compile(r'[^a-zA-Z@.]+').split(text)
    for w in words:
        if w not in stops and len(w)>2:
            word.append(w.lower())
    return word
```

```python
clean_text=clean_data(text)
```

```python
clean_text
```

```python
dict=[["beto","o'rourke"],["ted","cruz"]]
```

```python
l=len(clean_text)
for i in range(l-2):
    #find the whole name mentioned in tweet 
    if dict[0][0] in clean_text[i]:
        dict[0][0]=clean_text[i]
        if dict[0][1] in clean_text[i+1]:
            clean_text.remove(clean_text[i+1])
            l=l-1
        else:
            clean_text[i]=dict[0][0]
    elif dict[1][0] in clean_text[i]:
        if dict[1][1] in clean_text[i+1]:
            clean_text.remove(clean_text[i+1])
            l=l-1
        else:
            clean_text[i]=dict[1][1]
```

```python
len(clean_text)
```

```python
fdist = FreqDist(clean_text)
fdist
```

```python
fre=[]
for key in fdist.keys():
    if fdist[key] in range(1,100):
        fre.append(key)
```

```python
for i in fre:
    if i in nltk.corpus.words.words():
        print(i)
```
